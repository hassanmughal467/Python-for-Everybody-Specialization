```python
import sqlite3
```

```python
conn=sqlite3.connect('emaildb_1.sqlite')
cur=conn.cursor() 
```

```python
cur.execute('DROP TABLE IF EXISTS Counts')
```

```python
cur.execute('CREATE TABLE Counts(org TEXT, count INTEGER)')
```

```python
fname=input("input filename: ")
f=open(fname)
for line in f:
    if not line.startswith('From:'):continue
    pieces=line.split()
    email=pieces[1]
    parts = email.split('@')
    org = parts[-1]
    print(org)
    cur.execute('SELECT count FROM Counts WHERE org=?',(org,)) #? is placeolder  And this is a way to basically make sure that we don't allow SQL injection.
    #Go Google SQL injection to get a sense of what that is,this is a placeholder in this SQL, that will ultimately be replaced by this.
    row=cur.fetchone()
    if row is None:
        cur.execute('INSERT INTO Counts(org,count) VALUES(?,1)',(org,))
    else:
        cur.execute('UPDATE Counts SET count=count+1 WHERE org=?',(org,))
conn.commit()
sqlstr='SELECT org,count FROM Counts ORDER BY  count DESC Limit 10'
for row in cur.execute(sqlstr):
    print(str(row[0]),row[1])
cur.close()
```
